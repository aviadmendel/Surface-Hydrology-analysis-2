#  Thornthwaite equation

The first method to estimate the it was developed by Thornthwaite not so far from today in 1948 which is an empirical method and it predict monthly evapotranspiration. This equation is usually attractive because of the fact it needs such a small amount of data to be collected. To be specified by having the monthly mean of temperature we will receive the monthly mean of it. It can be very helpful in places without lots of sensors which will collect different kind of data.

Because we know it is affected of many more parameters we should be a bit skeptical with the results that we have but maybe even if they close enough to the actual Et we should be satisfied. Even Thornthwaite himself doubted his own equation by saying “The mathematical development is far from satisfactory. It is empirical. … The chief obstacle at present to the development of a rational equation is the lack of understanding of why potential ET corresponding to a given temperature is not the same everywhere”
It would be reasonable to say that the method is not that far from real life for the area that the empirical tests was taken and if maybe we could collect data for different kinds of climate and areas we could have different equations that based only on temperature which will give better results.

When looking at the equation, we can see the effect of the temperature on Et and we expect that the higher will go with the temperature the difference in it will be more significant because of the exponent in the equation.


$$
E = 16\left[ \frac{10\,T^\text{monthly mean}}{I} \right]^a,
$$

where

$$
I = \sum_{i=1}^{12} \left[ \frac{T_i^\text{monthly mean}}{5} \right]^{1.514},
$$

and

$$
\begin{align}
a &= 6.75\times 10^{-7}I^3 \\
   &- 7.71\times 10^{-5}I^2 \nonumber\\
   &+ 1.792\times 10^{-2}I \nonumber\\
   &+ 0.49239 \nonumber
\end{align}
$$

 - $E$ is the monthly potential ET (mm)
 - $T_\text{monthly mean}$ is the mean monthly temperature in °C
 - $I$ is a heat index
 - $a$ is a location-dependent coefficient

# Penman equation

Moving on to our next method to estimate Et, we will now talk about Penman equation which developed by chance or not in 1948 exactly the same year as Thornthwaite equation. Apparently lots of things happened that year not only for the Jewish people, but also in the aEt equation world.

Penman equation not like Thornthwaite is not empirical but a theory based to predict evaporative flux. 
Not like Thornthwaite equation the model requires lots of data which is not always easy to collect and some of it cost a lot of money especially when trying to use accurate equipment.

The data will need to find Et using the Penman equation will require temperature, radiation, wind speed, elevation above sea level, vapor-pressure deficit, and heat flux density to the ground.
Using all the equations below, we will now start to find each one of the variables sepretaly eventually puting them all in the Penman equation and finding estimated Et.

Using all the equations below, we will now start to find each one of the variables separately eventually putting them all in the Penman equation and finding estimated it.


### Penman full equation
$$
E = \frac{1}{\lambda}\left[ \frac{\Delta}{\Delta+\gamma}Q_{ne}+ \frac{\gamma}{\Delta+\gamma}E_A \right],
$$

where $Q_n$ is the available energy flux density

$$
Q_n = R_n - G,
$$

and $E_A$ is the drying power of the air

$$
E_A = 6.43\cdot f(u)\cdot\text{VPD}.
$$


$$
\gamma = \frac{c_p\, P}{\lambda\cdot MW_\text{ratio}}
$$
  
$$
P = 101.3-0.01055 H
$$

$$
\lambda = 2.501 - 2.361\times 10^{-3}\,T
$$

* $MW_\text{ratio}=0.622$: ratio molecular weight of water vapor/dry air
* $P$: atmospheric pressure (kPa). Can be either measured or inferred from station height above sea level (m).
* $\lambda$: latent heat of water vaporization (MJ kg$^{-1}$)








$$
R_n = (1-\alpha)R_s\!\! \downarrow -R_b \!\! \uparrow,
$$

where $\alpha$ (dimensionless) is the albedo.
The net outgoing thermal radiation $R_b$ is given by

$$
R_b = \left( a\frac{R_s}{R_{so}+b} \right)R_{bo},
$$

where $R_{so}$ is the solar radiation on a cloudless day, and it depends on latitude and day of the year.
$R_{bo}$ is given by

$$
R_{bo} = \epsilon\, \sigma\, T^4_{Kelvin},
$$

where $\sigma=4.903\times 10^{-9}$ MJ m$^{-2}$ d$^{-1}$ K$^{-4}$, and $\epsilon$ is net net emissivity:

$$
\epsilon=-0.02+0.261 \exp\left(-7.77\times10^{-4}T_{Celcius}^2\right).
$$

The parameters $a$ and $b$ are determined for the climate of the area:
* $a=1.0$, $b=0.0$ for humid areas,
* $a=1.2$, $b=-0.2$ for arid areas,
* $a=1.1$, $b=-0.1$ for semihumid areas.

$$
G = 4.2\frac{T_{i+1}-T_{i-1}}{\Delta t}
$$

$$
\text{VPD} = e_s - e_d.
$$

For temperatures ranging from 0 to 50 °C, the saturation vapor pressure can be calculated with

$$
e_s = \exp \left[ \frac{16.78\, T -116.9}{T+237.3} \right],
$$

and the actual vapor pressure is given by

$$
e_d = e_s \frac{RH}{100},
$$











$$
\Delta = \frac{\text{d} e_s}{\text{d}T} = e_s(T)\cdot \frac{4098.79}{(T+237.3)^2}.
$$









$$
f(u) = 0.26(1.0 + 0.54\, u_2)
$$








The various components of the equations above are:

  $$
    \Delta = 0.200 \cdot (0.00738\,T + 0.8072)^7 - 0.000116
  $$


  $$
    f_e(u) = 1.0 + 0.53\, u_2
  $$

 
where $\Delta t$ is the time *in days* between midpoints of time periods $i+1$ and $i−1$, and $T$ is the air temperature (°C).


 - $\Delta$: slope of the saturation water vapor pressure curve (kPa °C$^{-1}$)
 - $\gamma$: psychrometric constant (kPA °C$^{-1}$)
 - $c_p=0.001013$: specific heat of water at constant pressure (MJ kg$^{-1}$ °C$^{-1}$)
 - $P$: atmospheric pressure (kPa)
 - $H$: elevation above sea level (m)
 - $\lambda$: latent heat of vaporization (MJ kg$^{-1}$)
 - $R_n$: net radiation (MJ m$^{-2} d^{-1}$)
 - $G$: heat flux density to the ground (MJ m$^{-2} d^{-1}$)
 - $u_{2}$: wind speed measured 1.5 m above ground (m s$^{-1}$)
 - $e_{s} - e_{d}$: vapor pressure deficit (kPa) 
 - $e_{s}$: saturation vapor pressure (kPa) 
 - $e_{d}$: actual vapor pressure (kPa) 

```python
# creating an array and starting to add columns of data - temperature, wind (1.5 meters) and RH

# Resampling hourly data over same day and taking mean, to obtain daily averages

## Yosemite
df_YOP = (df_YO['AIR_TEMPERATURE'].resample('D')
                 .mean()
                 .to_frame()
                     )
df_YOP.rename(columns = {'AIR_TEMPERATURE':'T'}, inplace = True)
df_YOP['U'] = (df_YO['WIND_1_5'].resample('D')
                               .mean()
              )
df_YOP['RH'] = (df_YO['RELATIVE_HUMIDITY'].resample('D')
                                           .mean()
               )

## Everglades city

df_ECP = (df_EC['AIR_TEMPERATURE'].resample('D')
                 .mean()
                 .to_frame()
                     )
df_ECP.rename(columns = {'AIR_TEMPERATURE':'T'}, inplace = True)
df_ECP['U'] = (df_EC['WIND_1_5'].resample('D')
                               .mean()
              )
df_ECP['RH'] = (df_EC['RELATIVE_HUMIDITY'].resample('D')
                                           .mean()
               )
```

```python
#functions for lambda, Delta and G - all depend on T which we already have and air pressure 

def lambda_latent_heat(T):
    """daily latent heat of vaporization (MJ/kg)"""
    return 2.501 - 2.361e-3*T

def Delta(T):
    """slope of saturation-vapor curve (kPa/°C)"""
    return 0.2000*(0.00738*T + 0.8072)**7 - 0.000116

def G(T):
    """heat flux density to the ground, G (MJ/m2/d)"""
    return 4.2*np.gradient(T.values)

# finding pressure - Yosemite height above sea level 1209
df_YOP['P'] = 101.3-0.01055*1209

cp = 0.001013  # (MJ kg−1 °C−1) 
df_YOP['lambda'] = lambda_latent_heat(df_YOP['T'])
df_YOP['Delta'] = Delta(df_YOP['T'])
df_YOP['G'] = G(df_YOP['T'])
df_YOP['gamma'] = (cp*df_YOP['P'])/(0.622*df_YOP['lambda'])
df_YOP['f_wind'] = 1.0 + 0.53 * df_YOP['U']

## Everglades city

# finding pressure - Everglades height above sea level 0
df_ECP['P'] = 101.3-0.01055*0

cp = 0.001013  # (MJ kg−1 °C−1) 
df_ECP['lambda'] = lambda_latent_heat(df_ECP['T'])
df_ECP['Delta'] = Delta(df_ECP['T'])
df_ECP['G'] = G(df_ECP['T'])
df_ECP['gamma'] = (cp*df_ECP['P'])/(0.622*df_ECP['lambda'])
df_ECP['f_wind'] = 1.0 + 0.53 * df_ECP['U']
```

### calculating Rn:

  $$
  R_{bo} = \epsilon\, \sigma\, T^4_{Kelvin},
  $$
  where
  $$
  \epsilon=-0.02+0.261 \exp\left(-7.77\times10^{-4}T_{Celcius}^2\right),
  $$  
  $$
  \sigma=4.903\times 10^{-9} \text{  MJ m$^{-2}$ d$^{-1}$ K$^{-4}$},
  $$
  and
  $$
  T_{Kelvin}=T_{Celcius}+273.15
  $$
* Calculate 
  $$
  R_b = \left( a\frac{R_s}{R_{so}+b} \right)R_{bo},
  $$
  where  
  - for humid areas, $a=1.0$ and $b=0$,  
  - for arid areas, $a=1.2$ and $b=-0.2$,  
  - for semihumid areas, $a=1.1$ and $b=-0.1$
* Finally, calculate 
  $$
  R_n = (1-\alpha)R_s\!\! \downarrow -R_b \!\! \uparrow,
  $$
  where  
  - $\alpha= 0.23$ for most green crops with a full cover
  - $\alpha= 0.04$ for fresh asphalt
  - $\alpha= 0.12$ for worn-out asphalt
  - $\alpha= 0.55$ for fresh concrete

```python
# creating a function using data which will collect evetually calculate Rn

def R(a,b,df_pen):
    sigma = 4.903e-9
    emissivity = -0.02 + 0.261 * np.exp(-7.77e-4 * df_pen['T']**2)
    
    # Rbo: net longwave radiation for clear skies, otherwise known as diffuse radiation or emitted radiation from the
    Rbo = emissivity*sigma*((df_pen['T']+273.15)**4)

    # net outgoing long-wave radiation (note: Rs/Rso = proportion of how clear the day is)
    # for humid areas, a=1.0 and b=0
    # for arid areas, a=1.2 and b=-0.2
    # for semihumid areas, a=1.1 and b=-0.1
    Rb = (a*df_pen['Rs_MJ']/df_pen['Rso_monthly'] + b)*Rbo   

    # α is the albedo, or short-wave reflectance (dimensionless) 
    alpha = 0.23

    # net radiation
    Rn = (1 - alpha) * df_pen['Rs_MJ'] - Rb   # (MJ/m2/d)
    df_pen['Rn'] = Rn
    
# since the data we have is in W/m^2 we will create a function to convert it to MJ/(m^2*d)
def convert_units(Rs):  
    return Rs*3600*24/(10**6)
```

```python
# Rso: mean solar radiation from a cloudless sky (based on latitude) -
# using the latitude for each location and a table of mean solar radiation 
# MJ/m2/d

##Yosemite
Rso_monthly_YO = np.array([2.51,5.99,13.82,22.32,
                           29.01,31.95,29.85,23.32,
                           15.78,8.5,3.64,1.55])

df_YOP["Rso_monthly"] = ""

# every day in the month will have the same values for Rso
for i in range(12):
    df_YOP.loc[df_YOP.index.month==(i+1), "Rso_monthly"] = Rso_monthly_YO[i]
df_YOP["Rs"]=df_YO["SOLAR_RADIATION"].resample('D').mean()

#using function to convert units
df_YOP['Rs_MJ']=df_YOP['Rs'].apply(convert_units)

#using function to calculate Rn
R(1.1,-0.1,df_YOP)

##everglades city
Rso_monthly_EC = np.array([17.46, 21.65, 25.96, 29.85,
                        32.11, 33.20, 32.66, 30.44,
                        26.67, 22.48, 18.30, 16.04])

df_ECP["Rso_monthly"] = ""
# every day in the month will have the same values for Rso
for i in range(12):
    df_ECP.loc[df_ECP.index.month==(i+1), "Rso_monthly"] = Rso_monthly_EC[i]
df_ECP["Rs"]=df_EC["SOLAR_RADIATION"].resample('D').mean()

#using function to convert units
df_ECP['Rs_MJ']=df_ECP['Rs'].apply(convert_units)

#using function to calculate Rn
R(1.1,-0.1,df_ECP)
```

### Calculating the vapor pressure deficit, VPD, using the following equations:

$$
e_d = e_s\cdot \frac{RH}{100}
$$

$$
e_s = \exp\left(\frac{16.78\,T-116.9}{T+237.3}\right)
$$

```python
# vapor pressure deficit = VPD
# creating a function to calculate es

def vp_sat(T):
    return np.exp((16.78*T - 116.9)/(T + 237.3)) 

## Yosemite
df_YOP['es'] = vp_sat(df_YOP['T'])
df_YOP['ed'] = df_YOP['es'] * df_YOP['RH'] / 100
df_YOP['VPD'] = df_YOP['es'] - df_YOP['ed']

## everglades city

df_ECP['es'] = vp_sat(df_ECP['T'])
df_ECP['ed'] = df_ECP['es'] * df_ECP['RH'] / 100
df_ECP['VPD'] = df_ECP['es'] - df_ECP['ed']
```

Now that all variables have been defined we can use the Penman equation: 

$$
E_{tp} = \frac{\Delta}{\Delta+\gamma}Q_{n}+ \frac{\gamma}{\Delta+\gamma}E_A
$$

$Q_n$ is the available energy flux density:

$$
Q_n = R_n - G,
$$

and $E_A$ is the drying power of the air:

$$
E_A = f_e(u)\cdot\text{VPD}
$$


```python
# creating a function that collect all the data we found and 
# using it in penman equation to find EP
def E_penman(df):
    T = df['T']
    Delta = df['Delta']
    gamma = df['gamma']
    Rn = df['Rn']
    G = df['G']
    EA = 6.43*df['f_wind'] * df['VPD']
    lambd = df['lambda']
    return ((Delta / (Delta + gamma))*(Rn - G) + ((gamma / (Delta + gamma))*EA)) / lambd

##Yosemite
df_YOP['E_penman'] = E_penman(df_YOP)
# since evaporation cant be negative we will convert all negative numbers to 0
df_penclean0_YO = df_YOP['E_penman']
df_penclean0_YO[df_penclean0_YO < 0] = 0
df_YOP['E_pen'] = df_penclean0_YO

##Everglades city
df_ECP['E_penman'] = E_penman(df_ECP)
# since evaporation cant be negative we will convert all negative numbers to 0
df_penclean0_EC = df_ECP['E_penman']
df_penclean0_EC[df_penclean0_EC < 0] = 0
df_ECP['E_pen'] = df_penclean0_EC
```

## The parameters affect on Penman equation
We will now try to assess the parameters that affect the differences between the two locations in Penmans results. We will focus on RH, T, VPD and Wind speed tough obviously all parameters affect and contribute the Et but I chose this few parameters because they showed some differences that match the differences we saw in the graphs.
Lets remember on  the equation and the relevant parameters we are going to talk about.

$$
E = \frac{1}{\lambda}\left[ \frac{\Delta}{\Delta+\gamma}Q_{ne}+ \frac{\gamma}{\Delta+\gamma}E_A \right],
$$

and $E_A$ is the drying power of the air

$$
E_A = 6.43\cdot f(u)\cdot\text{VPD}.
$$

$$
\lambda = 2.501 - 2.361\times 10^{-3}\,T
$$


* $\lambda$: latent heat of water vaporization (MJ kg$^{-1}$)

$$
\text{VPD} = e_s - e_d.
$$

For temperatures ranging from 0 to 50 °C, the saturation vapor pressure can be calculated with

$$
e_s = \exp \left[ \frac{16.78\, T -116.9}{T+237.3} \right],
$$

and the actual vapor pressure is given by

$$
e_d = e_s \frac{RH}{100},
$$


$$
f(u) = 0.26(1.0 + 0.54\, u_2)
$$


The various components of the equations above are:

  $$
    \Delta = 0.200 \cdot (0.00738\,T + 0.8072)^7 - 0.000116
  $$


  $$
    f_e(u) = 1.0 + 0.53\, u_2
  $$


 - $\Delta$: slope of the saturation water vapor pressure curve (kPa °C$^{-1}$)
 - $\gamma$: psychrometric constant (kPA °C$^{-1}$)
 - $c_p=0.001013$: specific heat of water at constant pressure (MJ kg$^{-1}$ °C$^{-1}$)
 - $\lambda$: latent heat of vaporization (MJ kg$^{-1}$)
 - $u_{2}$: wind speed measured 1.5 m above ground (m s$^{-1}$)
 - $e_{s} - e_{d}$: vapor pressure deficit (kPa) 
 - $e_{s}$: saturation vapor pressure (kPa) 
 - $e_{d}$: actual vapor pressure (kPa) 
 
### RH and VPD
The two graphs on the left shows the relative humidity and VPD (Vapour pressure deficit) tells us two things, first of all, the fact that in general as we expected the RH near the ocean in Everglades is higher than Yosemite and constant all year long and very high around 75% and higher (sounds sweaty) while in Yosemite the maximum is around 65% and during the summer goes down to 35%.
It is quite straight forward from the equations, but even easier to see in the two graphs the opposite influence of the two - high RH means lower VPD.
Looking at the equations Et is higher when> Ea is higher > VPD is higher > ed is lower> RH is lower. It is all a big chain that affected by few parameters and a big component of them is RH. That's a first good explanation of why during the summer when the RH in Yosemite is lower the Et is higher while the RH in Everglades is very high and the Et is lower.
Not getting into all details the simple explanation that we can understand is that high VPD means that the air is quite dry and "demand" more water from the ground and plants to balance it.
